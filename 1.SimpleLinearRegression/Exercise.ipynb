```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.columns = df.columns.str.replace('$', '')
df.columns = df.columns.str.replace('(', '')
df.columns = df.columns.str.replace(')', '')
```

```python
df.head()
```

```python
df.columns = df.columns.str.replace(' ','_')
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income')
plt.scatter(df.year,df.per_capita_income_US,color='red',marker='+')
```

```python
new_df = df.drop('per_capita_income_US',axis='columns')
```

```python
new_df
```

```python
income = df.per_capita_income_US
```

```python
income
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df,income)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
i = reg.predict(new_df)
```

```python
i
```

```python
new_df['income']=i
new_df
```

```python
new_df.to_csv("output.csv")
```

```python
plt.xlabel('year',fontsize=20)
plt.ylabel('income',fontsize=20)
plt.scatter(df.year,df.per_capita_income_US,color='red',marker='+')
plt.plot(df.year,i,color='blue')
```
